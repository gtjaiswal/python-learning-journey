The "Price Auditor" Challenge ðŸ’°
To audit these prices, we need to "drill down" through the layers. Let's try to reach the price 30 (the very last book in the second store).

Direct Access: Write the code to access that 30 using three indices: book_chain[?][?][?].

The "Grand Total" Traversal: Using the "nested loop" logic we discussed (3 loops deep), write a small script that print()s every single price in the entire chain, one by one.

How many for loops will you need to write to reach those individual prices, and what would your direct access indices be for the number 30?

```python
book_chain = [
    [[10, 12], [15, 18]], # Store 1
    [[20, 22], [25, 30]]  # Store 2
]
needed_book = book_chain[1][1][1]
print(needed_book)
grand_total=0
for store in book_chain:
    for section in store:
        for book in section:
            grand_total+=book
print(grand_total)
```

Step 3: The "Surgical" Slice ðŸ”ª
Now let's look at Multi-dimensional Slicing. This is different from a comprehension because instead of "looping," we are defining a coordinate range to "cut" out a piece of the array.

In Python (especially when using libraries like NumPy, though it works with basic lists too if you're careful), we use the colon : to mean "everything in this dimension."

Let's say we want to check the first sensor of every shelf in every rack.

Rack: All of them (:)

Shelf: All of them (:)

Sensor: Just the first one (0)

With standard Python lists, we usually have to use a comprehension to do this "cross-section" slice: [rack[shelf][0] for rack in server_farm for shelf in range(3)]

Try this: If you wanted to create a 2D list that contains only the first shelf from every rack, how would you write that comprehension?

Hint: You want to keep the rack loop, but instead of looping through shelves, just pick index

```python
# How would you write a single line of code to extract all the sensor data from just the middle Rack (Rack 1)? Hint: You only need one index to grab that entire "slice."

server_farm = [
    [[22, 25], [21, 23], [20, 24]], # Rack 0
    [[30, 35], [42, 45], [38, 40]], # Rack 1 (The "Hot" Rack)
    [[24, 26], [22, 25], [21, 23]] # Rack 2
]

[sensor for shelf in server_farm[1] for sensor in shelf if int(sensor) > 40]

# the first shelf from every rack
[rack[0]  for rack in server_farm]

# first shelf from first rack
[rack[0] for rack in server_farm[1]]

# How would you write that comprehension to get a single flat list of all 9 shelves from the server_farm?
shelves = [shelf for rack in server_farm for shelf in rack]
shelves
```
