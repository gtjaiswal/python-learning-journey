```python
txn_list = [2, 4, 3, 6, 8, 5, 9,11, 2,4]
target = 8
txn_dict = {}
for i, num in enumerate(txn_list):
    if num <= target:
        txn_dict[i]=num
        print(txn_dict)
# sorted_keys=sorted(txn_dict)
values = list(txn_dict.values())
print(type(values))
final_list=[]
for j,num1 in enumerate(values):
    sublist = values[j+1:]
    for k,num2 in enumerate(sublist, start=j+1):
        if num1 + num2 == target:
            final_list.append((txn_dict[num1],txn_dict[num2]))
            # final_list.append(txn_dict[num2])
print(final_list)
```
