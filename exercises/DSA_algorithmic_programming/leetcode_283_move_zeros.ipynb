# 283. Move Zeroes

Given an integer array nums, move all 0's to the end of it while maintaining the relative order of the non-zero elements.

Note that you must do this in-place without making a copy of the array.



Example 1:

Input: nums = [0,1,0,3,12]
Output: [1,3,12,0,0]
Example 2:

Input: nums = [0]
Output: [0]


Constraints:

1 <= nums.length <= 104
-231 <= nums[i] <= 231 - 1


```python
from typing import List
nums = [0,1,0,3,12]
class Solution:
    def moveZeroes(self, nums: List[int]) -> None:
        print(nums)
        for num in nums:
            if num==0:
                nums.remove(0)
                nums.append(0)
        print(nums)

    def moveZeroes_twoPointers(self, nums: List[int]) -> None:
        print(nums)
        i=j=0
        for i,j in range (len(nums)):
            if not nums[j]==0:
                nums[i]=nums[j]
                nums[j]=0
                i=j
        print(nums)

sol = Solution()
# sol.moveZeroes(nums)
sol.moveZeroes_twoPointers(nums)
```
