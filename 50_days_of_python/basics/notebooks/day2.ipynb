```python
from pathlib import Path
from collections import Counter
from operator import itemgetter
import string

# form the file path
def form_file_path():
    project_root = Path.cwd().resolve().parent
    data_dir = project_root/"data"
    txt_file_name="day1_task_word_frequency_input.txt"
    txt_file=data_dir/txt_file_name
    return txt_file

txt_file = form_file_path()
print(txt_file)
```

```python
# read the file
def read_file_contents(txt_file):
    with txt_file.open() as file:
        text = file.read().strip()
        # print(f"file content : \n {text}") if text else print("file is empty")
    return text

text = read_file_contents(txt_file)
print(text)   
```

```python
# clean the file contents
def clean_file_contents(text):
    if text:
        text=text.lower()
        text = "".join(ch for ch in text if ch not in string.punctuation)
        text= "".join(ch for ch in text if (ch in string.ascii_letters) or (ch in string.whitespace))
    return text
text = clean_file_contents(text)
print(text)  
```
